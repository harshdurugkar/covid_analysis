```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
india = pd.read_csv('covid_india.csv')
india=india.fillna(int(0))
india.head()
```

```python
total_case = max(india['total_cases'])
total_death = int(max(india['total_deaths']))
population = 1393409033
single_vac = int(max(india['people_vaccinated']))
fully_vac = int(max(india['people_fully_vaccinated']))


print("Total Case is",total_case)
print("Toatl Death is",total_death)
print("Total Population is",population)
print("People with Single Vaccinated is",single_vac)
print("People with Fully Vaccinated is",fully_vac)
```

```python
# This is Pie Chart of Total Covid Cases in India
fig = plt.figure()
ax = fig.add_axes([1,1,1.2,1.6])
label = ["Total Population" , "Total Cases"]
myexplode = [0.2,0.2]
color = ["#53f241","#e4f70c"]
value = [population,total_case]
plt.pie(value, labels = label,autopct='%1.2f%%', explode = myexplode, shadow = True, colors = color)
plt.legend(title = "Pie Chart Show Total Covid Case In India")
plt.show()
```

```python
# This is Pie Chart of Total Covid Death in India
fig = plt.figure()
ax = fig.add_axes([1,1,1.2,1.2])
label = ["Total Population" , "Total Death"]
myexplode = [0.2,0.2]
color = ["#53f241","#ed461c"]
value = [population,total_death]
plt.pie(value, labels = label,autopct='%1.2f%%', explode = myexplode, shadow = True, colors = color)
plt.legend(title = "Pie Chart Show Total Covid Death In India")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1.2,1.4])
label = ["Total Population" , "Single Vaccinated"]
myexplode = [0.1,0]
color = ["#53f241","#0cdcf7"]
value = [population,single_vac]
plt.pie(value, labels = label,autopct='%1.2f%%', explode = myexplode, shadow = True, colors = color)
plt.legend(title = "Pie Chart Show People with Single Vaccinated In India")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1.2,1.4])
label = ["Total Population" , "Fully Vaccinated"]
myexplode = [0.1,0]
color = ["#53f241","#e600ff"]
value = [population,fully_vac]
plt.pie(value, labels = label,autopct='%1.2f%%', explode = myexplode, shadow = True, colors = color)
plt.legend(title = "Pie Chart Show People with fully Vaccinated In India")
plt.show()
```
